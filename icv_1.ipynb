```python
#

import cv2
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import signal

def test1() :
    """"""
    z = np.zeros((2,4) , np.uint8)
    #print(type(z))
    #print(z)
    o = np.ones((2,4) , np.int32)
    #print(o)
    m = np.array([[4,12,3,1],[10,12,14,29]] , np.float32)
    #print(m)
    print(m[0,0])


def calcGrayHist(image) :
    rows , cols = image.shape
    grayHist = np.zeros([256],np.uint64)
    for r in range(rows) :
        for c in range(cols) :
            grayHist[image[r][c]] += 1 
    return grayHist

def pltGrayHist() :
    image = cv2.imread("D:/dieLuftDerFreiheitWeht/profile/creatist/cv/data/m1.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    gh = calcGrayHist(image)
    x_range = range(256)
    plt.plot(x_range,gh,linewidth=2,c='black')
    y_maxValue = np.max(gh)
    plt.axis( [ 0 , 255 , 0 , y_maxValue])
    plt.xlabel('gray level')
    plt.ylabel("number of pixels")
    plt.show()

def pltGrayHistWtihCVDef() :
    image = cv2.imread("D:/dieLuftDerFreiheitWeht/profile/creatist/cv/data/m1.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    rows , cols = image.shape 
    pixelSequence = image.reshape([rows*cols,])
    h,b,p = plt.hist(pixelSequence , 256 , facecolor='black' , histtype='bar')
    plt.xlabel('gl')
    plt.ylabel('num of pixel')
    y_maxValue = np.max(h)
    plt.axis([0,255,0,y_maxValue])
    plt.show()
    
def test2() :
    image = cv2.imread("D:/dieLuftDerFreiheitWeht/profile/creatist/cv/data/m1.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    a = 1.1
    o = float(a) * image
    o[o>255] = 255 
    o = np.round(o)
    o = o.astype(np.uint8)
    cv2.imshow("I" , image)
    cv2.imshow("O" , o)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

def histNormalize() :
    image = cv2.imread("D:/m0.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    Imax = np.max(image)
    Imin = np.min(image)
    Omin , Omax = 0 , 255
    a = float((Omax-Omin)/(Imax-Imin))
    b = Omin - a * Imin
    O = a * image + b 
    O = O.astype(np.uint8)
    cv2.imshow("I" , image)
    cv2.imshow("O" , O)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def gammaT() :
    image = cv2.imread("D:/m0.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    fI = image/255.0
    gamma = 1.5 
    O = np.power(fI,gamma) # float result
    cv2.imshow("I" , image)
    cv2.imshow("O" , O)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def gaussBlur(image , sigma , H , W, _boundary='fill' , _fillvalue=0) :
    gaussKernel_x = cv2.getGaussianKernel(sigma , W , cv2.CV_64F)
    gaussKernel_x = np.transpose(gaussKernel_x)
    gaussBlur_x = signal.convolve2d(image , gaussKernel_x , mode='same' ,
                                    boundary=_boundary,fillvalue=_fillvalue)
    
    gaussKernel_y = cv2.getGaussianKernel(sigma , H , cv2.CV_64F)
    gaussBlur_xy = signal.convolve2d(gaussBlur_x , gaussKernel_y,mode='same' , 
                                    boundary = _boundary , fillvalue=_fillvalue)
    return gaussBlur_xy    

def gaussBlur_test() :
    image = cv2.imread("D:/m0.jpg" , \
                       cv2.IMREAD_GRAYSCALE)
    cv2.imshow("image" , image)
    blurImage = gaussBlur(image , 8 , 51,51 , 'symm')
    blurImage = np.round(blurImage)
    blurImage = blurImage.astype(np.uint8)
    cv2.imshow("image blur" , blurImage)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
if __name__ == "__main__" :
    print(" cv comer")
    #pltGrayHist()
    #pltGrayHistWtihCVDef()
    #test2()
    #gammaT()
    gaussBlur_test()
    print(" end")
```
