```python
# %load kmeans2.py
#

import numpy as np 
import matplotlib.pyplot as plt 

import kmeans
import p_tool


class Kmeans_2split :

    
    
    def __init__( self, data_set, k ) :
        self.k = k
        self.data_set = data_set  
        self.m = np.shape(self.data_set)[0]  
        self.n = np.shape(self.data_set)[1]  #column number 
        self.clusterDist = np.mat(np.zeros((self.m,2))) # init dist
        self.center0 = np.mean(data_set , axis=0).tolist()[0] 
        #the center about all of the point as one cluster , all of eles's mean set to center 
        self.cenlist = [self.center0 ] 
        self.clusterCenter = None  # use to draw test

                
        
    def cluster(self) :
        """ 2 split kmeans cluster """
        for j in range(self.m) :
            self.clusterDist[j,1] = \
                np.linalg.norm(self.center0-data_set[j,:])**2
        # the distance between all of elements and they's center(center0)  
        while( len(self.cenlist) < self.k) :
            lowestSSE = np.inf 
            #sse
            for i in range( len(self.cenlist)) :
                pst = data_set[ np.nonzero(self.clusterDist[:,0].A==i)[0],:]
                centroidMat , splitClustAss = kmeans.Kmeans(pst , 2 ).cluster() 
                # to use 1 generation kmeans to split to 2 class 
                sseSplit = sum(splitClustAss[:,1])
                sseNotSplit = sum(self.clusterDist[np.nonzero(self.clusterDist[:,0].A != i)[0] , 1])
                # sum of squre of error  
                # with it to compare which class best 
                if (sseNotSplit + sseSplit) < lowestSSE :
                    bestCentToSplit = i 
                    bestNewCents = centroidMat
                    bestClustAss = splitClustAss.copy()
                    lowestSSE = sseNotSplit + sseSplit
                #update best cluster 
            bestClustAss[np.nonzero(bestClustAss[:,0].A == 1)[0],0] = len(self.cenlist)
            bestClustAss[np.nonzero(bestClustAss[:,0].A == 0)[0],0] = bestCentToSplit
            # update new cluster's center number 
            self.cenlist[bestCentToSplit] = bestNewCents[0,:].tolist()[0]
            self.cenlist.append(bestNewCents[1,:].tolist()[0])
            # update list of center of cluster 
            self.clusterDist[np.nonzero(self.clusterDist[:,0].A==bestCentToSplit)[0],:] = bestClustAss
            # update cluster which was selected (possess best min of sse)


            
    #************************************************************************
    # tool part , present here just for test with have not brain
    def _color_cluster(self) : 
        idx2 = 0 
        indexmem = self.clusterDist[:,0]
        for idx1 in range(self.m) :
            if indexmem[idx1] == 0 :
                plt.scatter(self.data_set[idx2,0] ,
                self.data_set[idx2, 1] , c='blue' , marker='o')
            elif indexmem[idx1] == 1 :
                plt.scatter(self.data_set[idx2,0] ,
                self.data_set[idx2, 1] , c='green' , marker='o')
            elif indexmem[idx1] == 2 :
                plt.scatter(self.data_set[idx2,0] ,
                self.data_set[idx2, 1] , c='red' , marker='o')
            idx2 += 1


    def _proto_data_set(self) :
        plt.scatter( self.data_set[:,0].tolist() , self.data_set[:,1].tolist() ,
        c='red' )
        plt.show() 


    def _drawCenter(self) :
        self.clusterCenter = np.mat(self.cenlist)
        plt.scatter( self.clusterCenter.T[0].tolist() , self.clusterCenter.T[1].tolist() ,
        s=20 , c='black' , marker = 'D')

    def _drawResult(self) :
        self._color_cluster()
        self._drawCenter()
        plt.show() 
    #*******************************************************************************



if __name__ == '__main__' :
    print(' test to dischotomy_kmeans ')
    data_set = p_tool.file2matrix("D:/dieLuftDerFreiheitWeht/profile/creatist/ml/imitate_autism/iris.csv" , ',')   
    k2 = Kmeans_2split(data_set , 3)
    k2.cluster()
    k2._drawResult()
```
